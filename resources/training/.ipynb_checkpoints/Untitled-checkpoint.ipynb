```python
import numpy as np
import tensorflow as tf
import pickle

tf.config.threading.set_inter_op_parallelism_threads(2)
```

```python
with open('./train_numbers.pkl', 'rb') as f:
    train_x = tf.convert_to_tensor(pickle.load(f)[:].reshape((-1, 28, 28, 1)), dtype=tf.float32)
with open('./train_labels.pkl', 'rb') as f:
    train_y = np.array(pickle.load(f)[:]) - 1
with open('./test_numbers.pkl', 'rb') as f:
    test_x = tf.convert_to_tensor(np.array(pickle.load(f)).reshape((-1, 28, 28, 1)), dtype=tf.float32)
with open('./test_labels.pkl', 'rb') as f:
    test_y = np.array(pickle.load(f)) - 1
    
# test_y = np.array([9,6,7,1,1,2,8,4,7,4,9,6,2,7,8,5,8,4,5,9,7,1,6,9,3,2,8,3,7,1,9,8,5,3,4,1,8,6]) -1
```

```python
# import matplotlib.pyplot as plt
# for i in range(10):
#     index = int(np.random.rand()*train_x.shape[0])
#     plt.imshow(train_x[index,:,:,0]*255)
#     plt.title(train_y[index])
#     plt.show()
```

```python
#model = tf.keras.models.Sequential([
#    tf.keras.layers.Flatten(input_shape=(28,28)),
#    tf.keras.layers.Dropout(0.2),
#    tf.keras.layers.Dense(128, activation='relu'),
#    tf.keras.layers.Dense(128, activation='relu'),
#    tf.keras.layers.Dropout(0.2),
#    tf.keras.layers.Dense(9)
#])

model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(16, (5, 5), padding='same', activation='relu', activity_regularizer=tf.keras.regularizers.l2(0.05)),
    tf.keras.layers.MaxPool2D(pool_size=(2, 2)),
    tf.keras.layers.Conv2D(32, (5, 5), padding='same', activation='relu', activity_regularizer=tf.keras.regularizers.l2(0.05)),
    tf.keras.layers.MaxPool2D(pool_size=(2, 2)),
    tf.keras.layers.Dropout(0.25),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation='relu', activity_regularizer=tf.keras.regularizers.l2(0.05)),
    tf.keras.layers.Dense(9)
])
```

```python
predictions = model(train_x).numpy()
tf.nn.softmax(predictions).numpy()
loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
```

```python
model.compile(optimizer='adam',
              loss=loss_fn,
              metrics=['accuracy'])
```

```python
model.fit(train_x, train_y, validation_data=(test_x, test_y), epochs=5)
```

```python
import matplotlib.pyplot as plt
predictions = np.argmax(model(test_x).numpy(), axis=1) + 1
k = 0
for i in range(predictions.shape[0]):
    if predictions[i] != test_y[i]+1:
        plt.imshow(test_x[i,:,:,0]*255)
        plt.title('Predicted: %d' % (predictions[i]))
        plt.show()
    if k > 10:
        break
    k += 1
```

```python
model.save('dense_digits')
```
